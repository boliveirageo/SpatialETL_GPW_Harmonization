```python
src = '../datasets/10_CropHarvest/labels.geojson'
df = gpd.read_file(src)
df
```

```python
dfpoint = df[df.geom_type == 'Point']
dfpoint.index = range(0,dfpoint.shape[0])
dfpoint['gpw_lulc_class'] = dfpoint['label'].apply(lambda x: classgpw(x))
dfpoint['reference_year'] = dfpoint['collection_date'].apply(lambda x: x.year)
dfpoint['dataset_name'] = 'Cropharvest'
dfpoint.rename(columns={'label':'original_lulc_class'},inplace=True)

dfpoint
```

```python
dfpoint.isna().sum()
```

```python
dfpoint
```

```python
dfpoint = dfpoint[dfpoint.original_lulc_class.notnull()]
```

```python
dfpointFinal = dfpoint[['lat','lon','dataset_name', 'reference_year', 'original_lulc_class', 'gpw_lulc_class']]
```

```python
crs = "EPSG:4326"
newdf = gpd.GeoDataFrame(dfpointFinal, geometry=gpd.points_from_xy(dfpointFinal.lon, dfpointFinal.lat),crs=crs)
newdf.plot()
```

```python
newdf['sample_weight'] = 1
newdf['observation'] = ''
newdf
```
